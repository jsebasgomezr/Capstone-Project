```python
import pandas as pd
```

```python
URL = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
Table = pd.read_html(URL)
```

```python
# Scrape the information Table from Wikipedia and read it into a DataFrame
postcode = Table[0]
postcode.head()
```

```python
# Clean up the DataFrame from 'Not Assigned' Boroughs
index = postcode[postcode['Borough']=='Not assigned'].index
postcode.drop(index, inplace=True)
postcode = postcode.reset_index(drop=True)
postcode.head(10)
```

```python
# Group the DataFrame's cells by their Postcode and Borough
df_toronto= postcode.groupby(by=['Postcode','Borough']).agg(lambda x: ','.join(x))
df_toronto.reset_index(level=['Postcode','Borough'], inplace=True)
df_toronto.head(10)
```

```python
# Replace 'Not Assigned' Neighborhoods for their correspondent Boroughs
df_Toronto=df_toronto[df_toronto.Borough != 'Not assigned']
df_Toronto.head(10)
```
